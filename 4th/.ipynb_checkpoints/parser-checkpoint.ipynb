```python
import urllib.request as urllib2
import urllib
import re
from bs4 import BeautifulSoup

def parse_node(graph, url):
    graph[url] = set() 
    responce = urllib2.urlopen(url)
    
    soup = BeautifulSoup(responce,"lxml")
    for link in soup.find_all('a', href=True):
        goal_link = urllib.parse.urljoin(url, link.get('href'))
        if (urllib.parse.urlparse(url).netloc == urllib.parse.urlparse(goal_link).netloc and
            urllib.parse.urlparse(url).scheme == urllib.parse.urlparse(goal_link).scheme):
            graph[url].add(goal_link)

url = 'http://ugrafmsh.ru/'
graph = dict()

parse_node(graph, url)

for link in sorted(graph[url])[:15]:
    print(link)
    
```

```python
import queue
import time

def create_graph(url, depth=100,loggin=False):
    graph = dict()
    qu = queue.Queue()
    qu.put((url,0))
    link, d = qu.get()
    while d < depth:
        if loggin:
            print("Parce %s" % link)
        if link in graph:
            continue
        parse_node(graph, link)
        for child in graph[link]:
            if child not in graph:
                qu.put((child, d+1))
        if qu.empty():
            break
        link, d = qu.get()
    return graph


graph = create_graph(url,depth=2, loggin=True)
```

```python
from igraph import Graph

vertexes = list(sorted([list(sorted(graph[v]))[i]for v in graph.keys() for i in range(len(graph[v])) ]))
if url not in vertexes:
    vertexes.append(url)
G = Graph(edges = [(vertexes.index(a), vertexes.index(v)) for v in graph.keys() for a in graph[v]])
```

```python
labels = vertexes
N=len(labels)
E=[e.tuple for e in G.es]
layt=G.layout('kk',  dim=3) 
type(layt)
```

```python
def page_rank_algo(graph, damping_factor=0.1, eps=0.0001):
    page_rank = dict()
    num_urls = len(graph)
    for url in graph:
        page_rank[url] = 1. / num_urls
    diff = eps * 10
    while diff > eps:
        new_pr = dict()
        for url in graph:
            new_pr[url] = damping_factor / num_urls
        for url in graph:
            n_children = len(graph[url])
            if n_children > 0:
                for child in graph[url]:
                    if child not in new_pr:
                        new_pr[child] = 0
                    new_pr[child] += (1 - damping_factor) * page_rank[url] / n_children
            else:
                new_pr[url] += (1 - damping_factor) * page_rank[url]
        diff = 0
        for url in graph:
            diff += abs(new_pr[url] - page_rank[url])
        page_rank = new_pr.copy()
    return page_rank

```

```python
pagrank = page_rank_algo(graph, damping_factor=0.05)
```

```python
for url in pagrank:
    print ('url: %s pr: %.2f%%' % (url, 100 * pagrank[url]))
```
