```python
import web_parser

url = 'https://ml-mipt.github.io/'
graph = dict()
```

```python
%%time

graph, vertexes, links = web_parser.create_graph(url, loggin=True)
```

```python
pagrank = web_parser.page_rank_algo_new(graph, eps=0.05)

for url in pagrank:
    print ('url: %s pr: %.2f%%' % (url, 100 * pagrank[url]))
```

```python
web_parser.plot(vertexes, links, pagrank)
```

На этом сайте структура более всего похожая на дерево, откуда становится очевидным, что листья будут иметь наибольший ранг, так как в них можно попасть из большинства страниц
